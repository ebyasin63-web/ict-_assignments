```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
def calculate_electricity_bill(units):
    total_bill = 0
    if units <= 100:
        total_bill = units * 5
    elif units <= 200:
        total_bill = (100 * 5) + ((units - 100) * 7)
    else:
        total_bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)
    return total_bill

units_consumed_str = input("Enter units consumed: ")
try:
    units_consumed = int(units_consumed_str)
    if units_consumed < 0:
        print("Units consumed cannot be negative. Please enter a valid number.")
    else:
        bill = calculate_electricity_bill(units_consumed)
        print(f"Total Bill Amount: ₹{bill}")
except ValueError:
    print("Invalid input. Please enter a whole number for units consumed.")
```

```python
def is_palindrome(text):
    # Remove spaces and convert to lowercase for a robust check
    processed_text = ''.join(filter(str.isalnum, text)).lower()
    return processed_text == processed_text[::-1]

# Get input from the user
user_input = input("Enter a word or phrase to be checked: ")

# Check if it's a palindrome and print the result
if is_palindrome(user_input):
    print("Given num/string is Palindrome")
else:
    print("Given num/string is not Palindrome")
```
