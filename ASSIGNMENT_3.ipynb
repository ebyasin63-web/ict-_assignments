### 3. Python program to insert a number to any position in a list (Get the number to be inserted and position from the user)

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

number_to_insert = 99
position_to_insert = 2

if 0 <= position_to_insert <= len(my_list):
    my_list.insert(position_to_insert, number_to_insert)
    print(f"List after inserting {number_to_insert} at position {position_to_insert}: {my_list}")
else:
    print(f"Invalid position {position_to_insert}. Position must be between 0 and {len(my_list)}.")
```

### 4. Write a program to rename a key `city` to `location` in the following dictionary.
`sample_dict = { "name": "Kelly", "age":25, "salary": 8000, "city": "New york"}`

```python
sample_dict = { "name": "Kelly", "age":25, "salary": 8000, "city": "New york"}
print(f"Original dictionary: {sample_dict}")

if 'city' in sample_dict:
    sample_dict['location'] = sample_dict.pop('city')

print(f"Dictionary after renaming 'city' to 'location': {sample_dict}")
```

### 5. Write a Python program to change Bradâ€™s salary to `8500` in the following dictionary.
`sample_dict = {'emp1': {'name': 'Jhon', 'salary': 7500}, 'emp2': {'name': 'Emma', 'salary': 8000}, 'emp3': {'name': 'Brad', 'salary': 500}}`

```python
sample_dict_nested = {
    'emp1': {'name': 'Jhon', 'salary': 7500},
    'emp2': {'name': 'Emma', 'salary': 8000},
    'emp3': {'name': 'Brad', 'salary': 500}
}
print(f"Original dictionary: {sample_dict_nested}")

for emp_id, emp_info in sample_dict_nested.items():
    if emp_info.get('name') == 'Brad':
        emp_info['salary'] = 8500
        break

print(f"Dictionary after updating Brad's salary: {sample_dict_nested}")
```

### 6. Write a program to copy elements `44` and `55` from the following tuple into a new tuple.
Given: `tuple1 = (11, 22, 33, 44, 55, 66)`

```python
tuple1 = (11, 22, 33, 44, 55, 66)
print(f"Original tuple: {tuple1}")


new_tuple_method1 = (tuple1[3], tuple1[4])
print(f"New tuple (Method 1): {new_tuple_method1}")

elements_to_copy = [44, 55]
new_tuple_method3 = tuple(item for item in tuple1 if item in elements_to_copy)
print(f"New tuple (Method 3): {new_tuple_method3}")
```

### 7. Counts the number of occurrences of item `50` from a tuple
`tuple1 = (50, 10, 60, 70, 50, 83, 567, 50, 81)`

```python
tuple_to_count = (50, 10, 60, 70, 50, 83, 567, 50, 81)
item_to_find = 50
print(f"Original tuple: {tuple_to_count}")

occurrences = tuple_to_count.count(item_to_find)

print(f"The number of occurrences of {item_to_find} in the tuple is: {occurrences}")
```
