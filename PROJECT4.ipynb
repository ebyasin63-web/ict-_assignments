```python
acc = {}

print("\n" + "*"*45)
print("*      MINI BANKING SYSTEM        *")
print("*         YASIN ASSIGNMENT         *")
print("*"*45)

while True:
    print("\n1.Create  2.Deposit  3.Withdraw")
    print("4.Transfer  5.Balance  6.Exit")
    ch = input("Enter choice: ")

    if ch == "1":
        a = input("Account No: ")
        b = float(input("Balance: "))
        if b < 500:
            print("Minimum balance 500"); continue
        acc[a] = b
        print("Account created")

    elif ch == "2":
        a = input("Account No: ")
        amt = float(input("Amount: "))
        if a in acc:
            acc[a] += amt
            print("Deposited")
        else:
            print("Invalid account")

    elif ch == "3":
        a = input("Account No: ")
        amt = float(input("Amount: "))
        if a in acc and acc[a] - amt >= 500:
            acc[a] -= amt
            print("Withdrawn")
        else:
            print("Denied")

    elif ch == "4":
        f = input("From: ")
        t = input("To: ")
        amt = float(input("Amount: "))
        if f in acc and t in acc and acc[f] - amt >= 500:
            acc[f] -= amt; acc[t] += amt
            print("Transferred")
        else:
            print("Denied")

    elif ch == "5":
        a = input("Account No: ")
        print("Balance:", acc.get(a, "Invalid"))

    elif ch == "6":
        print("Thank you"); break
```

```python
# Personal Expense Tracker

from datetime import datetime

# Expense list: each expense is a dictionary
expenses = []

# Budget limits for categories (example)
budget_limits = {
    "Food": 500,
    "Transport": 200,
    "Entertainment": 150,
    "Utilities": 300,
    "Others": 100
}

def log_expense():
    """Log a new expense"""
    date_str = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category (Food, Transport, Entertainment, Utilities, Others): ")
    amount = float(input("Enter amount: "))

    # Validate date
    try:
        date = datetime.strptime(date_str, "%Y-%m-%d")
    except ValueError:
        print("Invalid date format! Use YYYY-MM-DD.")
        return

    # Store expense
    expenses.append({"date": date, "category": category, "amount": amount})
    print(f"Expense added: {date_str}, {category}, ${amount:.2f}")

def monthly_summary(month, year):
    """Generate summary for a given month and year"""
    total = 0
    category_totals = {}

    for expense in expenses:
        if expense['date'].month == month and expense['date'].year == year:
            total += expense['amount']
            category_totals[expense['category']] = category_totals.get(expense['category'], 0) + expense['amount']

    print(f"\n--- Monthly Summary for {year}-{month:02d} ---")
    print(f"Total Spending: ${total:.2f}")
    print("Category-wise Spending:")
    for cat, amt in category_totals.items():
        print(f"  {cat}: ${amt:.2f}")
        # Budget breach alert
        if cat in budget_limits and amt > budget_limits[cat]:
            print(f"  ALERT: Budget exceeded for {cat}!")

def analyze_spending():
    """Analyze high-spending patterns"""
    day_totals = {}
    category_totals = {}

    for expense in expenses:
        day = expense['date'].strftime("%Y-%m-%d")
        day_totals[day] = day_totals.get(day, 0) + expense['amount']
        category_totals[expense['category']] = category_totals.get(expense['category'], 0) + expense['amount']

    # Find highest expense day
    if day_totals:
        highest_day = max(day_totals, key=lambda d: day_totals[d])
        print(f"\nHighest Expense Day: {highest_day} - ${day_totals[highest_day]:.2f}")

    # Find dominant spending category
    if category_totals:
        dominant_category = max(category_totals, key=lambda c: category_totals[c])
        print(f"Dominant Spending Category: {dominant_category} - ${category_totals[dominant_category]:.2f}")

def main():
    while True:
        print("\nPersonal Expense Tracker")
        print("1. Log Expense")
        print("2. View Monthly Summary")
        print("3. Analyze Spending")
        print("4. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            log_expense()
        elif choice == '2':
            month = int(input("Enter month (1-12): "))
            year = int(input("Enter year (e.g., 2026): "))
            monthly_summary(month, year)
        elif choice == '3':
            analyze_spending()
        elif choice == '4':
            print("Exiting Expense Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

# Run the tracker
if __name__ == "__main__":
    main()
```
